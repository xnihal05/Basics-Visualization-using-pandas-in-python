```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
#creating a plot comparing the gender ratio according to stream
plt.plot(df["stream"],df["gender"], color="Blue")
```

```python
#Finding total students gender ratio in college in pie graph with percentage
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', title="Student's Gender in college", label ="")
```

```python
#Finding total gender ratio in bar graph form
df['gender'].value_counts().plot(kind='bar', color=["Blue","Pink","yellow"], title="Total Gender Ratio")
```

```python
#Total Stream Students ratio 
df['stream'].value_counts().plot(kind='pie', label="", title="Total Stream Sudents", autopct="%1.1f%%")
```

```python
#Creating mask for cs/it data --> going in stream --> takking cs/it stream only
mask = df['stream'].str.contains('CS/IT', case= False , na=False, regex=False)
```

```python
mask
```

```python
csit_data = df[mask]
```

```python
csit_data
```

```python
#Gender counts in specific cs/it department
csit_data['gender'].value_counts().plot(kind='pie',autopct='%1.1f%%',title='Gender Counts in CS/IT department', label="")
```

```python
csit_data['gender'].value_counts().plot(kind='bar', color=['blue','pink','yellow'], title='Gender Ratio Graphs of CS/IT', xlabel= 'Gender', ylabel='Total number of Students')
```

```python
#Doing similar for MBA data
MBA_data = df[df['degree_type']== 'MBA']
MBA_gender= MBA_data['gender'].value_counts()
plt.pie(
    MBA_gender,
    labels = MBA_gender.index,
    autopct= "%1.1f%%"
)
plt.title= 'MBA Gender Ratio'
plt.show
```
